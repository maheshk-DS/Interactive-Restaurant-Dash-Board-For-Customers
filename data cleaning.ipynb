```python
del df['url']
del df['address']
del df['phone']
```

```python
df.info()
```

```python
#Replacing restaurants with their ratings given as New to NAN and dropping them finally 
df['rate'] = df['rate'].replace('NEW',np.NaN)
df['rate'] = df['rate'].replace('-',np.NaN)
df.dropna(how = 'any', inplace = True)
```

```python
df.info()
```

```python
df['rate'] = df.loc[:,'rate'].replace('[ ]','',regex = True)
df['rate'] = df['rate'].astype(str)
df['rate'] = df['rate'].apply(lambda r: r.replace('/5',''))
df['rate'] = df['rate'].apply(lambda r: float(r))
```

```python
df.info()
```

```python
#Conversion of Cost from String to Integer
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',','')
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(int)
```
